```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_table = pd.read_csv('results.csv')
results_table
```

```python
# Create a figure and axis
fig, ax = plt.subplots()
# Hide axes
ax.axis('off')
# Create a table from the DataFrame
table = ax.table(cellText=results_table.values, colLabels=results_table.columns, cellLoc='left', loc='left', fontsize=16)
# Save the figure as a PNG image
fig.savefig('results_table.png')
# Display the table
plt.show()
table
```
